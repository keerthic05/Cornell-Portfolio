```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score
```

```python
# testing/training data
y = df['Freedom to make life choices']
X = df.drop(columns=['country','Freedom to make life choices'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
# building Linear regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_lr_pred = lr_model.predict(X_test)
lr_rmse = mean_squared_error(y_test, y_lr_pred)
lr_r2 = r2_score(y_test, y_lr_pred)
print('[LR] Root Mean Squared Error: {0}'.format(lr_rmse))
print('[LR] R2: {0}'.format(lr_r2))
```

```python
#plotting our model
# regression lineplot
plt.figure(figsize=(10,6))
plt.scatter(y_test, y_lr_pred, color='blue', label='Predicted vs. Actual')

plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect fit line')

plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Linear Regression: Actual vs Predicted Freedom to make life choices')
plt.show()
```
